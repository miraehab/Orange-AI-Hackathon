```python
import pandas as pd
import numpy as np
```

```python
persons = pd.read_csv('./titanic/train.csv')
```

<h3>The shape of the data</h3>

```python
persons.shape
```

<h3>know the number of male and female on the titanic </h3>

```python
persons.Sex.value_counts()
```

```python
per_male = (persons.Sex.value_counts()['male']/persons.shape[0])*100
print("the percentage of male on titanic is ", per_male, '%')
```

```python
per_female = (persons.Sex.value_counts()['female']/persons.shape[0])*100
print("the percentage of male on titanic is ", per_female, '%')
```

<h3> Persons with age greater than or equal 60</h3>

```python
old_persons = persons[persons.Age >= 60]
old_persons
```

```python
old_persons.shape[0]
```

<h3>Old persons that survived</h3>

```python
old_persons_survived = old_persons[old_persons.Survived == 1]
old_persons_survived
```

```python
old_persons_survived.shape[0]
```

```python
old_survived_per = (old_persons_survived.shape[0]/old_persons.shape[0])*100
print("Percentage of survived old persons from all old persons ", old_survived_per, '%')
```

<h3>Average of the age of Survived persons</h3>

```python
avrg_surv_age = persons[persons.Survived == 1].Age.mean()
round(avrg_surv_age)
```
